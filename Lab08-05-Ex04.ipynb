## คำถาม

1. Field ในภาษา C# คืออะไร


*คำตอบ*

Field คือ ตัวแปรที่ถูกประกาศภายในคลาสหรือโครงสร้างที่ใช้เก็บข้อมูลหรือสถานะของอ็อบเจ็กต์ในภาษา C#.

2. Field มีหน้าที่อะไรใน class 


*คำตอบ*

Field ในคลาสมีหน้าที่ในการเก็บข้อมูลหรือสถานะของอ็อบเจ็กต์แต่ละตัว ซึ่งจะใช้ภายในคลาสเพื่อเก็บและจัดการข้อมูลที่เกี่ยวข้องกับอ็อบเจ็กต์นั้น ๆ

3. การประกาศ Field ทำได้อย่างไร

*คำตอบ*

การประกาศ Field ใน C# สามารถทำได้โดยการกำหนดชื่อฟิลด์, ประเภทข้อมูลของมัน และสามารถกำหนดค่าเริ่มต้นให้กับฟิลด์ได้ด้วย. ฟิลด์สามารถเข้าถึงได้ตามระดับการเข้าถึงที่กำหนด (เช่น public, private, เป็นต้น).

4. Implicit fields initialization  คืออะไร

*คำตอบ*

Implicit fields initialization หมายถึง การที่ฟิลด์ (Field) ในคลาสได้รับค่าพื้นฐาน (default value) โดยอัตโนมัติเมื่อไม่กำหนดค่าเริ่มต้นให้กับฟิลด์นั้น ๆ ซึ่งใน C# เมื่อคุณประกาศฟิลด์ในคลาสที่ไม่ได้กำหนดค่าเริ่มต้นให้ มันจะได้รับค่าเริ่มต้น (default value) ตามประเภทข้อมูลที่ใช้

5. Explicit field initialization คืออะไร



*คำตอบ*

Explicit field initialization หมายถึง การกำหนดค่าเริ่มต้นให้กับฟิลด์ในคลาสอย่างชัดเจน โดยที่คุณจะระบุค่าของฟิลด์นั้น ๆ เมื่อประกาศฟิลด์ในคลาส ตัวอย่างเช่น การกำหนดค่าพื้นฐานที่ไม่ใช่ค่าดีฟอลต์ (default value) ที่ภาษา C# กำหนดให้


การทำ explicit field initialization คือการกำหนดค่าเริ่มต้นให้กับฟิลด์ทันทีที่ประกาศฟิลด์ในคลาส ซึ่งจะทำให้ฟิลด์มีค่าเริ่มต้นที่คุณตั้งใจตั้งแต่แรก

6. Implicit fields initialization  และ Explicit field initialization แตกต่างกันอย่างไร



*คำตอบ*

Implicit field initialization: ฟิลด์จะได้รับค่าพื้นฐานจาก C# โดยอัตโนมัติ เช่น 0 สำหรับ int หรือ null สำหรับ string หากไม่ได้กำหนดค่าเริ่มต้นเอง


Explicit field initialization: คุณกำหนดค่าเริ่มต้นให้ฟิลด์เอง เช่น กำหนด F1 = 10 หรือ F2 = "Hello" โดยตรงในการประกาศฟิลด์


การใช้ Explicit initialization จะช่วยให้ฟิลด์มีค่าเริ่มต้นที่คุณกำหนดเอง ซึ่งแตกต่างจาก Implicit ที่ใช้ค่าพื้นฐานที่ C# กำหนดให้โดยอัตโนมัติ.
